## Matriz de confusão:

    A matriz de confusão é especialmente útil quando se trata de problemas de classificação binária, onde o modelo tenta prever se uma instância pertence a uma das duas classes possíveis.

    Vamos entender os elementos básicos de uma matriz de confusão com base em uma classificação binária:

    Verdadeiros Positivos (VP): São os casos em que o modelo previu corretamente a classe positiva (1) e a verdade é que a instância pertence à classe positiva.

    Verdadeiros Negativos (VN): São os casos em que o modelo previu corretamente a classe negativa (0) e a verdade é que a instância pertence à classe negativa.

    Falsos Positivos (FP): Também conhecidos como "erros do tipo I", são os casos em que o modelo previu incorretamente a classe positiva, mas a verdade é que a instância pertence à classe negativa.

    Falsos Negativos (FN): Também conhecidos como "erros do tipo II", são os casos em que o modelo previu incorretamente a classe negativa, mas a verdade é que a instância pertence à classe positiva.

```python
from sklearn.metrics import confusion_matrix
print('Knn')
print(confusion_matrix(y_teste, predito_knn))

print('Bnb')
print(confusion_matrix(y_teste, predito_BNb))

print('ArvoreDecisao')
print(confusion_matrix(y_teste, predito_ArvoreDecisao))

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Supondo que você já tenha definido as variáveis predito_knn, predito_BNb e predito_ArvoreDecisao
# e y_teste com as previsões e rótulos verdadeiros, respectivamente.

# Matrizes de confusão
cm_knn = confusion_matrix(y_teste, predito_knn)
cm_bnb = confusion_matrix(y_teste, predito_BNb)
cm_arvore = confusion_matrix(y_teste, predito_ArvoreDecisao)

# Lista de matrizes de confusão
matrizes_confusao = [cm_knn, cm_bnb, cm_arvore]

# Nomes dos modelos
nomes_modelos = ['KNN', 'BNB', 'Árvore de Decisão']

# Configurar o subplot
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Criar mapas de calor para cada matriz de confusão
for ax, matriz_confusao, nome_modelo in zip(axes, matrizes_confusao, nomes_modelos):
    sns.heatmap(matriz_confusao, annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_title(f'Matriz de Confusão - {nome_modelo}')
    ax.set_xlabel('Predito')
    ax.set_ylabel('Verdadeiro')

# Ajustar layout
plt.tight_layout()
plt.show()
```

## Acurácia
### Mede o total que o modelo acertou:
    Ela representa a proporção de predições corretas em relação ao total de predições feitas pelo modelo.

    Matematicamente, se representarmos:

    TP (True Positives): o número de exemplos positivos corretamente classificados como positivos.

    TN (True Negatives): o número de exemplos negativos corretamente classificados como negativos.

    FP (False Positives): o número de exemplos negativos erroneamente classificados como positivos.

    FN (False Negatives): o número de exemplos positivos erroneamente classificados como negativos.
    A acurácia pode ser calculada por:

    Acurácia= TP+TN / TP+TN+FP+FN

    Em palavras simples, a acurácia responde à pergunta: "De todas as predições feitas pelo modelo, quantas estão corretas?". Embora a acurácia seja uma métrica fácil de entender, ela pode não ser a melhor escolha em todos os casos, especialmente em situações onde as classes estão desbalanceadas.

    A acurácia é utilizada quando queremos saber a performance geral do modelo, isto é, dentre todas as classificações realizadas, quantas foram classificadas corretamente pelo modelo.

```python
from sklearn.metrics import accuracy_score

print('Knn')
print(accuracy_score(y_teste, predito_knn))

print('Bnb')
print(accuracy_score(y_teste, predito_BNb))

print('ArvoreDecisao')
print(accuracy_score(y_teste, predito_ArvoreDecisao))
```

## Precisão
### Quantos foram calculados de forma correta:
    Mede a proporção de instâncias positivas previstas corretamente em relação ao total de instâncias positivas previstas pelo modelo. Em outras palavras, a precisão responde à pergunta: "De todas as instâncias que meu modelo classificou como positivas, quantas estão corretas?".
    Precisão = True Positives (TP) / True Positives (TP) + False Positives (FP)

    A precisão é utilizada quando queremos saber dentre todas as classificações de classe Positivo realizadas pelo modelo, quantas estão corretas.


```python
from sklearn.metrics import precision_score

print('Knn')
print(precision_score(y_teste, predito_knn))

print('Bnb')
print(precision_score(y_teste, predito_BNb))

print('ArvoreDecisao')
print(precision_score(y_teste, predito_ArvoreDecisao))
```

## Recall
### Quantos foram calculados de forma correta:
    Enquanto a precisão se concentra nas instâncias previstas como positivas, o recall mede a proporção de instâncias positivas que foram corretamente identificadas pelo modelo em relação ao total de instâncias positivas verdadeiras.
    Recall = Instancias positivas corretamente classificadas / Total de instancias que são realmente positivas
    Recall = True Positives (TP) / True Positives (TP) + False Negatives (FN)
    O recall é útil quando o custo dos falsos negativos é alto
    A precisão é relevante quando o foco está em minimizar os falsos positivos.
    O recall é relevante quando o objetivo é minimizar os falsos negativos.

```python
from sklearn.metrics import recall_score

print('Knn')
print(recall_score(y_teste, predito_knn))

print('Bnb')
print(recall_score(y_teste, predito_BNb))

print('ArvoreDecisao')
print(recall_score(y_teste, predito_ArvoreDecisao))
```

## F1 - Score:
### É uma média harmônica entre o recall e a precisão:
    O recall e a precisão são grandezas inversamente proporcionais, ou seja, à medida que vai aumentando o recall, por consequência irá diminuir a precisão e vice-versa.
    F1_score = 2 * (precisão * recall)/(precisão + recall)

```python
from sklearn.metrics import f1_score

print('Knn')
print(f1_score(y_teste, predito_knn))

print('Bnb')
print(f1_score(y_teste, predito_BNb))

print('ArvoreDecisao')
print(f1_score(y_teste, predito_ArvoreDecisao))
```

## Curva ROC e AUC:
### Curva ROC (Receiver Operating Characteristic):
    A Curva ROC é um gráfico que representa a relação entre a Taxa de Falsos Positivos (FPR) e a Taxa de Verdadeiros Positivos (TPR) para diferentes valores de limiar de probabilidade.
    O eixo X da curva representa o FPR (Taxa de Falsos Positivos), e o eixo Y representa o TPR (Taxa de Verdadeiros Positivos).
    A linha de referência diagonal (linha pontilhada) representa o desempenho aleatório, onde a capacidade preditiva do modelo é equivalente a uma escolha aleatória.
### AUC-ROC (Área sob a Curva ROC):
    A AUC-ROC é a medida da área sob a Curva ROC. Quanto maior a AUC, melhor o desempenho do modelo.
    A AUC-ROC varia de 0 a 1, onde 0 indica um desempenho ruim (pior que uma escolha aleatória) e 1 indica um desempenho perfeito.
    Uma AUC-ROC de 0,5 indica desempenho aleatório, enquanto uma AUC-ROC acima de 0,5 sugere que o modelo está superando uma escolha aleatória.
### Interpretação:
    Desempenho Aleatório: Se a curva ROC do modelo estiver próxima à diagonal de referência (linha pontilhada) e a AUC-ROC estiver próxima de 0,5, isso sugere que o modelo não está fornecendo informações discriminativas e está se comportando como uma escolha aleatória.

    Bom Desempenho: Quanto mais a curva ROC se afasta da linha de referência diagonal para o canto superior esquerdo, melhor é o desempenho do modelo. A AUC-ROC também será maior em casos de bom desempenho.

    Perfeição: Um modelo perfeito teria uma curva ROC que atingisse o canto superior esquerdo (TPR = 1, FPR = 0), resultando em uma AUC-ROC de 1.

### Avaliação dos Modelos:
    Comparação: Ao comparar modelos, o que tiver uma curva ROC mais próxima do canto superior esquerdo e uma AUC-ROC maior é geralmente considerado melhor.

    Trade-off: A escolha do modelo ideal pode depender do trade-off entre FPR e TPR, dependendo do contexto específico do problema. Por exemplo, em algumas situações, é mais importante minimizar os falsos positivos, enquanto em outras, maximizar os verdadeiros positivos é crucial.

    Ajuste do Limiar de Probabilidade: A Curva ROC pode ajudar a entender como o ajuste do limiar de probabilidade afeta o desempenho do modelo em diferentes aspectos.

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from itertools import cycle
import numpy as np

# Transformar rótulos em formato binário (necessário para a curva ROC)
y_teste_bin = label_binarize(y_teste, classes=np.unique(y_teste))

# Função para criar a curva ROC e calcular a AUC
def plot_roc_curve(model_name, y_test_bin, y_score, ax):
    fpr, tpr, _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
    roc_auc = auc(fpr, tpr)

    ax.plot(fpr, tpr, lw=2, label='{} (AUC = {:.2f})'.format(model_name, roc_auc))
    ax.plot([0, 1], [0, 1], linestyle='--', color='gray')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('Taxa de Falsos Positivos (FPR)')
    ax.set_ylabel('Taxa de Verdadeiros Positivos (TPR)')

# Criar a figura
fig, ax = plt.subplots()

# KNN
plot_roc_curve('KNN', y_teste_bin, predito_knn, ax)

# BNB
plot_roc_curve('BNB', y_teste_bin, predito_BNb, ax)

# Árvore de Decisão
plot_roc_curve('Árvore de Decisão', y_teste_bin, predito_ArvoreDecisao, ax)

# Adicionar legenda
ax.legend(loc="lower right")

# Mostrar a figura
plt.title('Curvas ROC para os Três Modelos')
plt.show()
```
