# RFE
### A eliminação recursiva de atributos (RFE, do inglês "Recursive Feature Elimination") é uma técnica utilizada em aprendizado de máquina para selecionar automaticamente as melhores características (ou atributos) em um conjunto de dados. O objetivo é melhorar o desempenho do modelo, removendo atributos irrelevantes ou menos importantes.

### O processo RFE geralmente funciona da seguinte maneira:

### Escolha do Modelo Inicial: Um modelo é treinado inicialmente com todas as características disponíveis no conjunto de dados.

### Atribuição de Importância: O algoritmo atribui importâncias ou pesos para cada atributo com base em sua contribuição para o desempenho do modelo.

### Eliminação dos Atributos Menos Importantes: Os atributos menos importantes são removidos do conjunto de dados.

### Retreinamento do Modelo: O modelo é retreinado com o conjunto de dados reduzido.

### Repetição do Processo: Os passos 2 a 4 são repetidos até que o número desejado de atributos seja alcançado ou até que o desempenho do modelo pare de melhorar significativamente.

```python
from sklearn.feature_selection import RFE

selecionador_rfe = RFE(estimator = classificador, n_features_to_select = 5, step = 1)
```

```python
from sklearn.feature_selection import RFE

SEED = 1234
random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v6,
                                                       diagnostico,
                                                       test_size = 0.3)

classificador = RandomForestClassifier(n_estimators=100, random_state = 1234)
classificador.fit(treino_x, treino_y)


selecionador_rfe = RFE(estimator = classificador, n_features_to_select = 5, step = 1)
selecionador_rfe.fit(treino_x, treino_y)
treino_rfe = selecionador_rfe.transform(treino_x)
teste_rfe = selecionador_rfe.transform(teste_x)
classificador.fit(treino_rfe, treino_y)

matriz_confusao = confusion_matrix(teste_y, classificador.predict(teste_rfe))
plt.figure(figsize = (10, 8))
sns.set(font_scale = 2)
sns.heatmap(matriz_confusao, annot = True, fmt = "d").set(xlabel = "Predição", ylabel = "Real")

print("Resultado da classificação %.2f%%" % (classificador.score(teste_rfe, teste_y)* 100))
```

# RFE CROSS VALIDATION

## Divide o banco de dados em blocos, e faz os testes nesses blocos, testando o melhor modelo

```python
from sklearn.feature_selection import RFECV

SEED = 1234
random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v6,
                                                       diagnostico,
                                                       test_size = 0.3)

classificador = RandomForestClassifier(n_estimators=100, random_state = 1234)
classificador.fit(treino_x, treino_y)


selecionador_rfecv = RFECV(estimator = classificador, cv = 5, scoring = "accuracy", step = 1)
selecionador_rfecv.fit(treino_x, treino_y)
treino_rfecv = selecionador_rfecv.transform(treino_x)
teste_rfecv = selecionador_rfecv.transform(teste_x)
classificador.fit(treino_rfecv, treino_y)

matriz_confusao = confusion_matrix(teste_y, classificador.predict(teste_rfecv))
plt.figure(figsize = (10, 8))
sns.set(font_scale = 2)
sns.heatmap(matriz_confusao, annot = True, fmt = "d").set(xlabel = "Predição", ylabel = "Real")

print("Resultado da classificação %.2f%%" % (classificador.score(teste_rfecv, teste_y)* 100))
```
